```python
import tensorflow as tf
import numpy as np
```

```python
x = tf.Variable(2.0)
y = tf.Variable(10.0)
```

```python
with tf.GradientTape(persistent = True) as tape:
    z = x**2 * tf.sin(y)
    a = z**3 + tf.exp(z)
    p = tf.math.log(a) + a**2

dp_by_dx = tape.gradient(p, x)
dp_by_dy = tape.gradient(p, y)

print("dp/dx -> (x=2, y=10) : ", dp_by_dx.numpy())
print("dp/dy -> (x=2, y=10) : ", dp_by_dy.numpy())
```
