```python
import pandas as pd
import numpy as np

file_path = "/content/drive/MyDrive/Colab Notebooks/KMC_student_dataset.xlsx"
df = pd.read_excel(file_path)

print(df)
```

```python
data_array = df.to_numpy()
data_array=data_array.astype(float)
print(data_array)
```

```python
normalized_data=np.zeros_like(data_array)
noOfRows = normalized_data.shape[0]
noOfColumns = normalized_data.shape[1]

for i in range(noOfColumns):
    xmin = np.min(data_array[:,i])
    xmax = np.max(data_array[:,i])
    normalized_data[:,i] = (data_array[:,i] - xmin) / (xmax-xmin)
print(normalized_data)
```

```python
k=3
noOfPoints = normalized_data.shape[0]
random_points = np.random.choice(noOfPoints, k, replace=False)

centroids = normalized_data[random_points]
print(centroids)

distances = np.zeros(shape=(noOfRows,k))
for i in range(k):
    distances[:,i] = np.sum(abs(normalized_data-centroids[i]),axis=1)

print(distances)

clusters = np.argmin(distances,axis = 1)
print(clusters)

new_clusters = np.zeros_like(clusters)

while True:
  new_centroids = np.zeros_like(centroids)
  print(new_centroids)
  for i in range(k):
    points = normalized_data[i==clusters]
    new_centroids[i] = points.mean(axis=0)

  print(new_centroids)

  for i in range(k):
    distances[:,i] = np.sum(abs(normalized_data - new_centroids[i]), axis = 1)

  print(distances)

  new_clusters = np.argmin(distances,axis = 1)
  print(new_clusters)

  if(np.array_equal(clusters, new_clusters)):
    break

  clusters = new_clusters

for i in range(k):
  points = normalized_data[i  == clusters]
  print("Cluster - " + str(i+1) + " : ")
  print(points)
```
